线性回归

```python
data_x = result[['maxdouer','mindouer','maxdouyou','mindouyou','maxdoupo','mindoupo','weahigh','wealow']]
```

```python
data_y = a1805[['结算价']]
```

```python
x_train,x_test,y_train,y_test =train_test_split(data_x, data_y, test_size=0.3)
```

```python
ss = StandardScaler()
x_train = ss.fit_transform(x_train)
x_test = ss.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)  

y_predict = lr.predict(x_test)
```

```python
import numpy as np
mse = np.average((y_predict-np.array(y_test))**2)
rmse = np.sqrt(mse)
print("均方误差平方和的平方根：",rmse)
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rcParams["font.sans-serif"] = [u"SimHei"]
mpl.rcParams["axes.unicode_minus"] = False

p = np.arange(len(x_test))
plt.figure(facecolor="w",figsize=(12,6))   #创建画布，facecolor为背景色，w是白色
plt.plot(p,y_test,"r-",linewidth = 2,label = "真实值")
plt.plot(p,y_predict,"g-",linewidth = 2,label = "预测值")
plt.legend(loc = "upper right")     #显示图例，设置图例的位置
plt.grid(b=True)
plt.show()
```
